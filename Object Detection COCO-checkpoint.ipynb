```python
import cv2 as cv
import numpy as np
from PIL import Image
```

```python
#setting up labels
label_path = "C:/users/imran/desktop/OD/labels2.txt"
with open(label_path) as file:
    labels = file.read().split('\n')
```

```python
# Paths to config file and pre-trained weights
config_file = "C:/users/imran/desktop/OD/ssd_mobilenet_v3_large_coco.pbtxt"
frozen_model = "C:/users/imran/desktop/OD/frozen_inference_graph.pb"
```

```python
# Loading a pretrained ssd(single shot detection) model.
model = cv.dnn_DetectionModel(frozen_model, config_file)
```

```python
# Modifying the model
# model.setInputSize(320,3120)
model.setInputSize(320,320)
model.setInputScale(1.0/127.5)   # 255/2 = 127.5
model.setInputMean((127.5, 127.5, 127.5))
model.setInputSwapRB(True)
```

```python
img = cv.imread("C:/users/imran/desktop/cupp.jpg")
```

```python
class_index, confidence, bbox = model.detect(img, confThreshold=0.5)
```

```python
class_index = np.array([51, 47])
```

```python
for class_ind, conf, boxes in zip(class_index.flatten(), confidence.flatten(), bbox):
    cv.rectangle(img, boxes, (0,255,0), 2)
    cv.putText(img, f"{labels[class_ind-1]}", (boxes[0]+10,boxes[1]+40), cv.FONT_HERSHEY_PLAIN, 1, (255,0,0), 2)
```

```python
cv.imshow("image", img)     
cv.imwrite("cups.jpg", img) # Saving image with bounding boxes on objects
cv.waitKey(0)
cv.destroyAllWindows()
```
